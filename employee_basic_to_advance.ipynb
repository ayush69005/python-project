```python
# Q28 Is higher salary associated with higher performance?
co_relation = Employee['monthly_salary'].corr(Employee['performance_rating'],method='spearman')

print(f"relation between the salary and performance is: {co_relation:.3f}")
```

```python
# visualization of the salary and performance
plt.scatter(Employee['monthly_salary'],Employee['performance_rating'], alpha=0.6)
plt.xlabel('Salary')
plt.ylabel('Performance_rating')
plt.title('Salary vs Performance')
plt.grid(True)
plt.show()
```

```python
# Q29 Do employees working longer hours perform better?
correlation_emp = Employee['working_hours_per_week'].corr(Employee['performance_rating'],method='spearman')
correlation_emp
print(f"realtion between working hour and performance rating : {correlation_emp:.3f}")
```

```python
# Q30 Does experience improve performance rating?
Exp_corr_perf = Employee['experience_years'].corr(Employee['performance_rating'], method='spearman')
Exp_corr_perf
print(f"relation between experience and performance rating: {Exp_corr_perf:.3f}")
```

```python
# Q31 Which department has the worst attrition problem?
dep_wst_attr = (
    Employee[Employee['attrition'] == 'Yes']['department']
    .value_counts()
    .reset_index(name='count')
    .rename(columns={'index': 'department'})
    .sort_values(by='count', ascending=False)
    .reset_index(drop=True)
)

dep_wst_attr
```

```python
# Q32 Are younger employees leaving more than older ones?

# creating the age group
Employee['age_groups']= pd.cut(
    Employee['age'],
    bins=[20, 30, 40, 50, 60, 70],
    labels=["20-30", "31-40", "41-50", "51-60", "61-70"]
)

# calculate attrition_rate
attrition_rates = (Employee.groupby('age_groups', observed=True)['attrition']
                   .value_counts(normalize=True)
                   .rename('attrition_rates')
                   .reset_index()
                   .round(2)

)

# calculating for employee who left
attrition_rates = attrition_rates[attrition_rates['attrition']==1]
attrition_rates
```

```python
# Q33 Compare performance rating across salary categories

# creating the salary category 
Employee['Salary_category']= (
    pd.cut(
        Employee['monthly_salary'],
        bins= [20000, 40000, 60000, 80000, 100000, 120000],
        labels= ['20000-40000', '40001-60000', '60001-80000', '80001-100000', '100001-120000']
    )
)

# performance rating across salary category
perfor_rating = (Employee.groupby('Salary_category',observed=True)['performance_rating']
                 .value_counts()
                 .reset_index()
                 .round(2)
                 )
perfor_rating
```

```python
# Verifying Q 33 
total_count= Employee['Salary_category'].value_counts()
total_count
```

```python
# Q34 Which department is most overworked (hours vs rating)?
Employee['overwork_score']= Employee['working_hours_per_week']/Employee['performance_rating']

Employee.groupby('department')['overwork_score'].mean().round(2).sort_values(ascending=False).reset_index()
```

```python
# Another way to solve Q34
(Employee.groupby('department')
 .agg(
     avg_working_hours = ('working_hours_per_week','mean'),
     Avg_performance_ratings = ('performance_rating','mean')
 )
 .round(2)
 .sort_values(by='Avg_performance_ratings' ,ascending=False)
 .reset_index()
 
 )
```
