# Apple Phone Sales Analysis

```python
# importing libraries:
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# Reading the csv file:
dataset= pd.read_csv("apple_products (1).csv")
dataset
```

```python
# cleaning of data:
# 1. missing values
Missing= dataset.isnull().sum()
Missing
```

```python
# 2. removing duplicate values:
Duplicate= dataset.duplicated()
Duplicate
```

```python
# Finding statistical data:
stat_data= dataset.describe()
stat_data
```

```python
# Sorting of data:
Highest_rated= dataset.sort_values(by='Star Rating',ascending=False).reset_index()
print(Highest_rated['Product Name'])
```

```python
# Finding to top N apple phone sold:
# let say N=10 
Highest_rated= Highest_rated.head(10)
print(Highest_rated['Product Name'])
```

```python
# visulization of data : number of rating by top 10 product .

sns.barplot(data=Highest_rated, x='Product Name', y='Number Of Ratings')
plt.title('Rating VS Product')
plt.xticks(rotation=90, ha='right')
plt.show()
```

```python
# plot a scatter graph to represent sales price by number of rating 

sns.regplot(data=dataset, 
                     x= 'Number Of Ratings', 
                       y=  'Sale Price', scatter=True, 
                         line_kws={"color": "red"})
plt.title('Sales Price by Number of rating')
plt.show()
```

```python
# plotting graph to represent discount percentage by number of rating:
Graph3= sns.regplot(data=dataset,
                     x= 'Number Of Ratings',
                       y=  'Discount Percentage',
                         scatter=True,
                           line_kws={"color": "red"})
plt.title('Discount percentage by Rating')
plt.show()
```
